# 활성화함수

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x들의 집합
xs = np.linspace(-5, 5, 501) # 0.1의 단위
```

## heaviside step function

```python
def heaviside(x):
    y = np.zeros_like(x)
    y[x>0] = 1  # x <= 0 이면 0, x > 1 이면 1
    return y
```

```python
plt.plot(xs, heaviside(xs))
plt.title('heaviside step func', fontsize = 16)
plt.grid(linestyle = ':')
plt.show()
```

## 시그모이드

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.plot(xs, sigmoid(xs))
plt.grid(linestyle = ':')
plt.title('Sigmoid', fontsize = 16)
plt.show()
```

## hyperbolic Tangent

```python
plt.plot(xs, np.tanh(xs))
plt.grid(linestyle = ':')
plt.title('hyperbolic Tangent', fontsize = 16)
plt.show()
```

## Relu(Lectified Linear Unit)

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.plot(xs,relu(xs))
plt.grid(linestyle = ':')
plt.title('Relu', fontsize = 16)
plt.show()
```

## leaky Lelu

```python
def leaky_relu(x):
    return np.maximum(0.01*x, x)
```

```python
plt.plot(xs,leaky_relu(xs))
plt.grid(linestyle = ':')
plt.title('leaky_relu', fontsize = 16)
plt.show()
```

## Soft Plus


```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
plt.plot(xs,softplus(xs))
plt.grid(linestyle = ':')
plt.title('softplus', fontsize = 16)
plt.show()
```

## Softmax

```python
def softmax(x):
    x = x - np.max(x)
    return np.exp(x) / np.sum(np.exp(x))
```

```python
yhat = np.array([-1, 1, 3, 3.5])
softmax(yhat)
```

```python
plt.plot(xs,softmax(xs))
plt.grid(linestyle = ':')
plt.title('softmax', fontsize = 16)
plt.show()
```
